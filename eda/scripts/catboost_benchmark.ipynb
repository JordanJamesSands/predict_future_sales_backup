```python
train2 = pd.read_csv('../../gen_data/train2.csv')
test1 = pd.read_csv('../../gen_data/test1.csv')
```

```python
train2.columns
```

```python
train2.head()
```

```python
test1.head()
```

```python
train_catboost = train2.drop(['Unnamed: 0','sales'],axis=1)
train_target = train2['sales']

test_catboost = test1.drop(['Unnamed: 0','ID'],axis=1)
test_catboost['date_block_num'] = 34
```

```python
train_catboost = train_catboost.drop(['shop_name','item_name','item_category_name'],axis=1)
test_catboost = test_catboost.drop(['shop_name','item_name','item_category_name'],axis=1)
```

```python
train_catboost.head()
```

```python
#reorder cols
test_catboost = test_catboost[['date_block_num','shop_id','item_id','item_category_id']]
test_catboost.head()
```

```python
model = catboost.CatBoostRegressor(loss_function='RMSE')
model.fit(train_catboost,train_target,cat_features=[1,2,3])
```

```python
preds = model.predict(test_catboost)
```

```python
preds
```

```python
test1[['shop_id','item_id','ID']].head()
```

```python
ans_catboost = test_catboost
ans_catboost['preds']=preds
test_catboost.head()
```

```python
catboost_submission = pd.merge(test_catboost,test1,on=['item_id','shop_id'])[['ID','preds']]
```

```python
catboost_submission.sort_values(by='ID',inplace=True)
```

```python
catboost_submission.columns = ['ID','item_cnt_month']
```

```python
catboost_submission.head()
```

```python
catboost_submission.to_csv('../../gen_data/catboost_submission.csv',index=False)
```
