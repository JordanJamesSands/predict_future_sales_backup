There's a lot of data in this project, by downcasting we can load it all into memory

```python
def downcast(df):
    float_cols = [col for col in df if df[col].dtype=='float64']
    int_cols = [col for col in df if df[col].dtype=='int64']
    df[float_cols] = df[float_cols].astype(np.float32)
    df[int_cols] = df[int_cols].astype(np.int32)
    return(df)
```

### Fixing price outliers

Let's look at the prices of items sold. Beginning by scaling them to a measure with standard deviation 1, call this `std_over_mean`

```python
mean_prices = train1.groupby('item_id').item_price.mean()
train1['mean_item_price'] = train1['item_id'].map(mean_prices)

median_prices = train1.groupby('item_id').item_price.median()
train1['median_item_price'] = train1['item_id'].map(median_prices)

std_prices = train1.groupby('item_id').item_price.std()
train1['std_item_price'] = train1['item_id'].map(std_prices)

train1['std_over_mean'] = (train1.item_price - train1.mean_item_price)/train1.std_item_price
```

Find prices with a high standard deviation from the mean price for that item

```python
item_counts = train1.groupby('item_id').item_price.count()
train1['item_count'] = train1.item_id.map(item_counts)
train1['threshold'] = -norm.ppf(1/(2*train1.item_count))
train1['toclip'] = (abs(train1.std_over_mean)>train1.threshold)
train1
```

## Data Leak

The mean of the public test set can be deduced via leaderboard probing, it only takes two constant solutions:

For $N$ = number of objects in prediction, For constant predictions $c_{0}$ and $c_{1}$, yeilding scores $S_{0}$ and $S_{1}$ respectively:

We know that: $S_{i} = \sqrt{\frac{1}{n}\sum_{j=1}^{j=N}{(y_j-c_i)^2}}$

$S_{i}^2 = \frac{1}{n}\sum{y_j} - \frac{2}{n}\sum{y_jc_i} + c_i^2$

and so

$S_0^2 - S_1^2 = \frac{2}{n}\sum{y_j(c_1-c_0)} + c_0^2 - c_1^2$

For $c_0 = 0$ and $c_1 = 1$ we get:

$S_0^2 - S_1^2 = \frac{2}{n}\sum{y_j} - 1$

and so

$\frac{S_0^2 - S_1^2 + 1}{2} = \frac{\sum{y_j}}{n}$

Plugging in data from probing:

$S_0 = 1.25011$ and $S_1 = 1.41241$

and so

$\frac{\sum{y_j}}{n} = \frac{1.25011^2 - 1.41241^2+1}{2}$

$\frac{\sum{y_j}}{n} = 0.2839365$

### Generate 0 Entries


We need to include entries where 0 sales were made for item/shop pairs in a month. If we did this for all shop item pairs possible, the dataset would be massive, so instead this is done for all item shop pairs **for each month**.

```python
def gen_zeros(train):
    months = range(train.date_block_num.min(),train.date_block_num.max()+1)
    to_pandas=[]
    print('Computing for month:',end=' ')
    for month in months:
        print(month,end=', ')
        subtrain = train[train.date_block_num==month].copy()
        all_shops = subtrain.shop_id.unique()
        all_items = subtrain.item_id.unique()
        pairs = product(all_shops,all_items)
        to_pandas.append([(month,x[0],x[1]) for x in pairs])
    
    train_filled = pd.DataFrame(np.vstack(to_pandas),columns=['date_block_num','shop_id','item_id'])
    
    #need to fill name vars
    #name_df = train[['shop_id','item_id','item_category_id','item_name','shop_name','item_category_name']]
    #train_filled = train_filled.merge(name_df)
    
    train_filled = train_filled.merge(train,how='left',on=['date_block_num','shop_id','item_id'])
    
    #Re-merge item_category_id and resort columns, this is computationally inefficient but saves time coding
    train_filled = train_filled.drop('item_category_id',axis=1).merge(items_data[['item_id','item_category_id']],on='item_id',how='left')
    cols = train_filled.columns.tolist()
    col_reseq = cols[0:3] + [cols[-1]] + cols[3:-1]
    train_filled = train_filled[col_reseq]
    #entries where item_cnt_month is supposed to be 0 are now created as NA
    
    #this line needs editing, dont fill verything with 0
    #what about new_item?
    train_filled.fillna(0,inplace=True)
    
    return train_filled
```

### Create Lag Values

Now generate historical features

```python
newvars = ['sum_item_sales_back_0','sum_shop_sales_back_0','item_cnt_month_back_0','sum_item_cat_sales_back_0','sum_city_back_0','sum_itemtype_back_0','sum_iteminfo_back_0']
```

```python
#make this more efficient?
def create_lag_train(train_filled):
    #gonna iterively copy a subset of the data, rename the date block and aome other cols then merge it back in dateback_gen = range(1,13)
    dateback_gen = range(1,DATEBACK_DIST+1)
    lag_train_filled = downcast(train_filled)
    del train_filled
    print('dateback=',end=' ')
    gc.collect()
    for dateback in dateback_gen: 
        #this line inneficient?
#       to_shift = lag_train_filled[['date_block_num','shop_id','item_id','sum_item_sales_back_0','sum_shop_sales_back_0','item_cnt_month_back_0','sum_item_cat_sales_back_0','sum_item_cat_shop_sales_back_0']].copy()
        to_shift = lag_train_filled[['date_block_num','shop_id','item_id']+newvars].copy()
        to_shift['date_block_num'] = to_shift.date_block_num + dateback
        newcols = ['date_block_num','shop_id','item_id'] + [x[0:-1]+str(dateback) for x in newvars]
        to_shift.columns = newcols
        #print(newcols)
        print(dateback,end=', ')
        lag_train_filled = lag_train_filled.merge(to_shift,on=['date_block_num','shop_id','item_id'],how='left').fillna(0)
        del to_shift
        gc.collect()
    
    #remove first DATEBACK_DIST many months
    lag_train_filled = lag_train_filled[lag_train_filled.date_block_num>=DATEBACK_DIST]
    return lag_train_filled
```

```python
start = time.time()
x_train_list = list(map(create_lag_train,x_train_list))
end = time.time()
print('Runtime: '+str((end-start)/60))
```

### Generate new item variable

It may be useful to know if an item has never been sold before, let's encode this as 1 for new item and 0 for already sold

```python
#start with sales data, figure out when each item was sold, 
    #merge this with train/test then transform it to 1/0
sales = pd.read_csv('gen_data/train1.csv')
def new_item_var(df):
    sales2 = sales.copy()
    first_sale_date = sales2.groupby('item_id').date_block_num.min()
    sales2['first_sale_date_block'] = sales2.item_id.map(first_sale_date)
    sales2 = sales2[['item_id','first_sale_date_block']]
    sales2 = sales2.drop_duplicates()
    df = df.merge(sales2,on='item_id',how='left')
    
    #handle nans for test data (which sales2 wont pick up)
    df.first_sale_date_block.fillna(999,inplace=True)    
    df['new_item'] = (df.first_sale_date_block >= df.date_block_num)*1

    return df
```

```python
x_train_list = list(map(new_item_var,x_train_list))
x_test_list = list(map(new_item_var,x_test_list))
```

# Month Features

Genereating features relating to the month in which data was aggregated

date_block_num==0 is January 2013

```python
def single_to_month(date_block_num):
    return ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec'][date_block_num % 12]
def to_month(df):
    df['month'] = pd.Series(map(single_to_month,df.date_block_num),index=df.index)
    return df
```

```python
def single_to_numdays(date_block_num):
    return [31,28,31,30,31,30,31,31,30,31,30,31][date_block_num % 12]
def to_numdays(df):
    df['numdays'] = pd.Series(map(single_to_numdays,df.date_block_num),index=df.index)
    return df
```

```python
x_train_list = list(map(to_month,x_train_list))
x_train_list = list(map(to_numdays,x_train_list))
```

```python
x_test_list = list(map(to_month,x_test_list))
x_test_list = list(map(to_numdays,x_test_list))
```

### Generate time specific market price data

Parsing some data related to the movement of price across **all** shops

```python
mean_price = sales.groupby(['date_block_num','item_id'],as_index=True).item_price.mean()
std_price = sales.groupby(['date_block_num','item_id'],as_index=True).item_price.std(ddof=1)
median_price = sales.groupby(['date_block_num','item_id'],as_index=True).item_price.median()

price_data = pd.concat([mean_price,std_price,median_price],axis=1)
price_data.columns = ['mean_prevmonth_item_price','std_prevmonth_item_price','median_prevmonth_item_price']
price_data.reset_index(inplace=True)
price_data['date_block_num'] += 1
```
