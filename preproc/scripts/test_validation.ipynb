```python
const_sols = [0.2,0.25,0.3,0.35,0.4,0.275,0.2875]
test_scores = [1.22032,1.21791, 1.21754,1.21922,1.22295,1.21747,1.21744]
```

```python
f = open('../../gen_data/master_validation_list.pickle','rb')
x_train_list , y_train_list , x_val_list, y_val_list = pickle.load(f)
```

```python
y_val_list[0].head()
```

```python
const_names = ['const_'+str(x) for x in const_sols]
validation_df = pd.DataFrame({'model': const_names, 'test_score': test_scores})
for i in range(33,20,-1):
    validation_df['valid_'+str(i)] = np.nan
validation_df
```

```python
n_folds = len(x_train_list)
for i in range(n_folds):
    loss_list=[]
    for const in const_sols:
        loss = np.sqrt(np.mean((const - y_val_list[i])**2))
        loss_list.append(loss)
    validation_df['valid_'+str(i+21)] = pd.Series(loss_list)
```
