### Study prices per item to identify outliers in price

```python
g = merged.groupby('item_id').item_price.mean()
merged['mean_item_price'] = merged['item_id'].map(g)

g = merged.groupby('item_id').item_price.std()
merged['std_item_price'] = merged['item_id'].map(g)

merged
```

```python
merged['std_over_mean'] = (merged.item_price - merged.mean_item_price )/merged.std_item_price
merged.sort_values('std_over_mean')
```
