```python
#find a factor to set as solution mean
SOL_MEAN = np.mean(newdf.item_cnt_month_back_1)
SOL_MEAN
```

```python
ALPHA=1

def validation(ALPHA,verbose=True):
    print('alpha='+str(ALPHA))
    SOLUTION_MEAN = 0.2556489262371615
    preds = test0.copy()

    preds['item_cnt_month']=0.0
    dateback_gen=range(1,34)


    for dateback in dateback_gen:
        
        item_cnt = newdf.iloc[:,dateback+3]
        if verbose:
            print('dateback = '+str(dateback),end='   ')
            print('colname = '+str(item_cnt.name))
        item_cnt = item_cnt.transform(lambda x: x*np.exp(-ALPHA*dateback))
        preds['item_cnt_month'] += item_cnt
    preds['item_cnt_month']=SOLUTION_MEAN/np.mean(preds.item_cnt_month) * preds.item_cnt_month

    return np.sqrt( np.mean( (preds.item_cnt_month - newdf.item_cnt_month_back_1)**2 ) )
```

```python
optimize.minimize_scalar(validation,tol=0.001)
#found 0.3723609580681796
```

```python
ALPHA=1

def retraining(ALPHA,verbose=True):
    print('alpha='+str(ALPHA))
    SOLUTION_MEAN = 0.2839853
    preds = test0.copy()

    preds['item_cnt_month']=0.0
    dateback_gen=range(1,35)


    for dateback in dateback_gen:
        
        item_cnt = newdf.iloc[:,dateback+2]
        if verbose:
            print('dateback = '+str(dateback),end='   ')
            print('colname = '+str(item_cnt.name))
        item_cnt = item_cnt.transform(lambda x: x*np.exp(-ALPHA*dateback))
        preds['item_cnt_month'] += item_cnt
    preds['item_cnt_month']=SOLUTION_MEAN/np.mean(preds.item_cnt_month) * preds.item_cnt_month

    return preds
```

```python
preds = retraining(0.3723609580681796)
```
