# Preprocessing

```python
def preproc_model3(x_train,x_test,y_train,y_test):
    x_train = [d.drop(['date_block_num','shop_id','item_id','item_category_id','month'],axis=1) for d in x_train]
    x_test =  [d.drop(['date_block_num','shop_id','item_id','item_category_id','month'],axis=1) for d in x_test]
    y_train = [d.clip(0,20) for d in y_train]
    y_test =  [d.clip(0,20) for d in y_test]
    return [x_train,x_test,y_train,y_test]
```

# Fitting

```python
model = xgb.XGBRegressor(
    seed=0,
    max_depth=2,
    learning_rate=0.5,
    n_estimators=1000,
    objective='reg:linear',
    nthread=7,
    min_child_weight=100,
    colsample_bytree=1,
    subsample=1
)
eval_set = [(x_train[0],y_train[0]),(x_test[0],y_test[0])]
```

```python
model.fit(
    verbose=True,
    X=x_train[0],
    y=y_train[0],
    eval_set=eval_set,
    early_stopping_rounds=10
)
```

```python
ax = xgb.plot_importance(model,height=0.8,)
print(ax.figure.set_size_inches(5,10))
```
